```python
import pandas as pd
import json
```

```python
data_df = pd.read_csv('../processed_data.csv')
code_map = json.load(open('../level.json', 'r', encoding='utf-8'))
```

```python
# 河南省 开封市 新区
```

```python
not_found = 0
def get_code(provice: str, city: str, district: str) -> str:
    level = None
    for prov in code_map:
        if prov['name'] == provice:
            if 'children' not in prov.keys():
                level = prov
                break
            for cit in prov['children']:
                if cit['name'] == city:
                    if 'children' not in cit.keys():
                        level = cit
                        break
                    for dist in cit['children']:
                        if dist['name'] == district:
                            level = dist
    if level is None:
        global not_found
        not_found += 1
        return '000000'
    return level['code']
```

```python
data_df['current_code'] = data_df.apply(lambda row: get_code(row['current_province'], row['current_city'], row['current_county']), axis=1)
```
